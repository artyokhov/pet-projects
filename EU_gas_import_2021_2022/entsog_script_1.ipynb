```python
import pandas as pd
import numpy as np
from entsog import EntsogPandasClient

pd.set_option('display.max_rows', 10)
pd.set_option('display.max_columns', 100)
pd.options.mode.chained_assignment = None

start = pd.Timestamp('20210101', tz='Europe/Brussels')
end = pd.Timestamp('20221231', tz='Europe/Brussels')

client = EntsogPandasClient()

# df_operations = client.query_operational_data_all(start = start, end = end, indicators = ['physical_flow'])
# df_operations.to_csv(r'/Users/artemohotnikov/Downloads/__dont_delete/gzprm/raw_data_2.csv', index=False)

df_operations = pd.read_csv(r'/Users/artemohotnikov/Documents/__code/gzprm/ready_2/raw_data_2.csv')
df_operations.head(5)
```

```python
df_operations.info()
```

```python
df_operations = df_operations.loc[(pd.isna(df_operations['value'])==False) & (df_operations['value']!=0) & (pd.isna(df_operations['point_type'])==False)]

df_operations['month'] = pd.to_datetime(df_operations['period_to'], utc=True).dt.month
df_operations['year'] = pd.to_datetime(df_operations['period_to'], utc=True).dt.year

df_operations = df_operations[['direction_key', 'id_point_type', 'point_type', 'operator_key', 'operator_label', 'period_to', 'month', 'year', 'point_key', 'point_label', 'tso_eic_code', 'tso_item_identifier', 'unit', 'value']]

df_operations.info()
```

```python
df_operations['value'] = pd.to_numeric(df_operations['value'], errors='coerce', downcast='float')

for col_name in ['id_point_type', 'month', 'year']:
    df_operations[col_name] = pd.to_numeric(df_operations[col_name], errors='coerce', downcast='integer')

df_operations.describe(include=[object])
```

```python
df_operations[['direction_key', 'point_type',
               'operator_key', 'operator_label',
               'point_key', 'point_label',
               'tso_eic_code', 'tso_item_identifier', 'unit', ]] = \
df_operations[['direction_key', 'point_type',
               'operator_key', 'operator_label',
               'point_key', 'point_label',
               'tso_eic_code', 'tso_item_identifier', 'unit', ]].astype('category')

df_operations.drop_duplicates(subset=None, keep='first', inplace=True, ignore_index=True)
df_operations['period_to'] = pd.to_datetime(df_operations['period_to'], utc=True)
df_operations = df_operations.rename(columns={'period_to': 'date'})

df_operations.info()
```

```python
df_operations[['value_in_bln_m3']] = df_operations.loc[:,['value']] / (10.5 * 1_000_000_000)
df_operations['value_in_bln_m3'] = np.around(df_operations.loc[:,'value_in_bln_m3'],3)

df_operations.loc[df_operations['direction_key']=='exit','value_in_bln_m3'] = \
df_operations.loc[df_operations['direction_key']=='exit','value_in_bln_m3'].apply(lambda x: x * -1)

df_operations.reset_index(drop=True, inplace=True)

df_operations.head(5)
```

```python
df_points = client.query_operator_point_directions()
df_points.head(3)
```

```python
df_operations = pd.merge(df_operations,
                         df_points[['point_key', 'direction_key', 'operator_key', 'adjacent_country']],
                         left_on = ['point_key','direction_key','operator_key'],
                         right_on = ['point_key','direction_key','operator_key'],
                         how='left')

df_operations.head(2)
```

```python
countries_dict = {'RU':'Russia',
                  'TR':'Turkey',
                  'NO':'Norway',
                  'LY':'Libya',
                  'TN':'Tunisia',
                  'PL':'Poland',
                  'SK':'Slovakia',
                  'HU':'Hungary',
                  'RO':'Romania',
                  'MA':'Morocco',
                  'UK':'United Kingdom', #check
                  'RO':'Romania',
                  'DE':'Germany',
                  'IT':'Italy',
                  'ES':'Spain',
                  'HR':'Croatia',
                  'NL':'Netherlands',
                  'CZ':'Czech Republic',
                  'BE':'Belgium',
                  'BG':'Bulgaria',
                  'FR':'France',
                  'LV':'Latvia',
                  'PT':'Portugal',
                  'GR': 'Greece',
                  'LU':'Luxembourg',
                  'EE':'Estonia',
                  'MK':'Macedonia',
                  'CH':'Switzerland',
                  'RS':'Serbia',
                  'SM':'San Marino',
                  'MD':'Moldova',
                  'UA':'Ukraine',
                  'AT':'Austria',
                  'SI':'Slovenia',
                  'DK':'Denmark',
                  'LT':'Lithuania',
                  'IE':'Ireland',
                  'FI':'Finland'}
```

```python
df_operations = df_operations[df_operations['adjacent_country']!='UK']  # отбрасываю строки с точками по UK

df_operations = df_operations[df_operations['tso_item_identifier']
                              .isin(['21YSF-EC-------Q', '21YEA-EC-------9', '21Z0000000000074', '21Z000000000081T', '21Z000000000088F'])==False]
```

```python
df_operations_pipe_import = df_operations.loc[df_operations['id_point_type']==1]  # Cross-Border Transmission IP between EU and Non-EU (import)

df_operations_pipe_import['adjacent_country'] \
                          .mask(df_operations_pipe_import['point_key'] \
                          .isin(['ITP-00094', 'ITP-00104', 'ITP-00092', \
                                 'ITP-00432' , 'ITP-00433', 'ITP-00434']), 'RU', inplace=True) # PL: Kondratki, Tieterowka, Wysokoje; UA: Uzhgorod-2, Uzhgorod-3, Uzhgorod-4

df_operations_pipe_import['adjacent_country'] \
                          .mask(df_operations_pipe_import['point_key'] == 'ITP-00089', 'PL', inplace=True) # to UA from PL: Drozdovichi (UA)

df_operations_pipe_import.head(3)
```

```python
# Для транзита через Укр в Польшу, Румынию, Эстонию/Латвию и Венгрию убрать exit, точки: ITP-00087, ITP-00493, ITP-10006, ITP-10008   + ужгород ITP-00117   + марокко ITP-00082

df_operations_pipe_import.loc[(df_operations_pipe_import['point_key']\
                              .isin(['ITP-00087', 'ITP-00493', 'ITP-10006', 'ITP-10008', 'ITP-00117', 'ITP-00082'])) & 
                              (df_operations_pipe_import['direction_key'] == 'exit'),
                              'value_in_bln_m3'] =\
df_operations_pipe_import.loc[df_operations_pipe_import['point_key']\
                              .isin(['ITP-00087', 'ITP-00493', 'ITP-10006', 'ITP-10008', 'ITP-00117', 'ITP-00082']),
                              'value_in_bln_m3'].apply(lambda x: x * 0)
```

```python
# Ямал-Европа 

df_operations_pipe_import = pd.concat([df_operations_pipe_import,
                                      df_operations[(df_operations['point_label'].str.contains('Mallnow')) &
                                                    (df_operations['direction_key']=='entry')]])
```

```python
df_operations_pipe_import['country_name'] = ''

for index, row in df_operations_pipe_import.iterrows():

    if row['adjacent_country'] in ['RU', 'TR', 'NO', 'LY'] and row['tso_item_identifier'] \
        not in ['58Z-00000015-S2M', '21Z0000000004758']:  # исключаю турецкий поток и азербайджан, ниже присваиваю страну
        df_operations_pipe_import.at[index,'country_name'] = countries_dict[row['adjacent_country']]

    elif row['adjacent_country'] in ['UA', 'BY', 'MD', 'EE', 'PL', 'SK', 'HU', 'RO']:
        df_operations_pipe_import.at[index,'country_name'] = 'Russia' # транзит

    elif row['adjacent_country'] == 'ES' and row['tso_item_identifier'] == '21Z0000000002131': # В Испанию труба из Марокко, а газ из Алжир, т.е. объемы отдать Алжиру
        df_operations_pipe_import.at[index,'country_name'] = 'Algeria'

    elif row['adjacent_country'] == 'MA' and row['tso_item_identifier'] == '21Z000000000045X': # В Испанию труба из Марокко, а газ из Алжир, т.е. объемы отдать Алжиру
        df_operations_pipe_import.at[index,'country_name'] = 'Algeria'

    elif row['adjacent_country'] == 'TN' and row['tso_item_identifier'] == '21Z000000000080V':  # В Италию по точке mazara газ из Алжира
        df_operations_pipe_import.at[index,'country_name'] = 'Algeria'
        
    elif row['adjacent_country'] == 'IT' and row['tso_item_identifier'] == '21Z000000000149L':  # В Италию по точке gela газ из Ливии
        df_operations_pipe_import.at[index,'country_name'] = 'Libya'

    elif row['tso_item_identifier'] == '58Z-00000015-S2M':
        df_operations_pipe_import.at[index,'country_name'] = 'Russia'  # это турецкий поток

    elif row['point_key'] == 'ITP-00274':
        df_operations_pipe_import.at[index,'country_name'] = 'Azerbaijan'  # Kipoi  это Азербайджан
```

```python
# тут вручную проверял дубли строк по объемам вручную, выявлены следующие наборы лейблов-дублей:

# Greifswald / OPAL (ITP-00251), Greifswald / NEL (ITP-00247)  - эти оставить
# Greifswald / GUD (ITP-00491),  Greifswald / LBTG (ITP-00454), Greifswald / Fluxys Deutschland (ITP-00297)  - эти убрать
# 'Dornum / NETRA (OGE)', 'Dornum / NETRA (GUD)'  21Z000000000053Y - убрать
#  Dornum GASPOOL   21Z0000000004855   - оставить
# 'Emden (EPT1) (Thyssengas)' - убрать
# 'Emden (EPT1) (GTS)', 'Emden (EPT1) (GUD)', 'Emden (EPT1) (OGE)' - оставить

list_to_except = ['Dornum / NETRA (OGE)', 'Dornum / NETRA (GUD)',
                  'Emden (EPT1) (Thyssengas)',
                  'Greifswald / GUD', 'Greifswald / LBTG', 'Greifswald / Fluxys Deutschland']  # эти точки нужно выкинуть

# здесь сортировкой по парам id и лейбл ручной поиск
pairs = df_operations_pipe_import[['point_label', 'tso_item_identifier']]
pairs.drop_duplicates(subset=None, keep='first', inplace=True, ignore_index=True)
pairs.sort_values('tso_item_identifier').head(20)
```

```python
df_operations_pipe_import = df_operations_pipe_import[df_operations_pipe_import['point_label'].isin(list_to_except)==False]  # убираю дубли

df_operations_pipe_import.loc[df_operations_pipe_import['point_key'] == 'ITP-00089','value_in_bln_m3'] = \
df_operations_pipe_import.loc[df_operations_pipe_import['point_key'] == 'ITP-00089','value_in_bln_m3'].apply(lambda x: x * -1)
# Drozdovichi (UA)  -Drozdowicze (PL)   ITP-00089  инвертировать exit и entry
```

```python
df_operations_pipe_import['country_name'].unique() # проверка на страны, чей газ ходит через импорт.точки
```

```python
# проверка в GWh для сверки с https://gasdashboard.entsog.eu/ по месяцам

df_operations_pipe_import['value_in_GWh'] = df_operations_pipe_import['value_in_bln_m3'] * 10.5 * 1_000

df_operations_pipe_import[df_operations_pipe_import['country_name']=='Norway'] \
                         [['point_key', 'country_name', 'year', 'month', 'value_in_GWh']] \
                         .groupby(by=['point_key', 'country_name', 'year', 'month']).sum()
```

```python
df_operations_lng_import = df_operations.loc[df_operations['id_point_type']==2]
df_operations_lng_import[['adjacent_country', 'country_name']] = 'LNG'

df_operations_lng_import[['year', 'value_in_bln_m3']].groupby(by='year').sum()

# проверка суммы импорта по годам всего в европу без UK по трубам
```

```python
df_operations_eu_production = df_operations.loc[df_operations['id_point_type']==3]  # Aggregated production point - TP

df_operations_eu_production.loc[pd.isna(df_operations_eu_production['adjacent_country']),'adjacent_country'] = \
df_operations_eu_production.loc[pd.isna(df_operations_eu_production['adjacent_country']),'operator_key'].apply(lambda x: x[:2])

df_operations_eu_production = df_operations_eu_production[df_operations_eu_production['direction_key']=='entry']
df_operations_eu_production.drop_duplicates(subset=None, keep='first', inplace=True, ignore_index=True) # по Италии дубли

df_operations_eu_production.loc[df_operations_eu_production['point_key'] \
                                .isin(['PRD-00202', 'PRD-00234', 'PRD-00240', 'PRD-00168', 'PRD-00200', 'PRD-00243']), 'value_in_bln_m3'] = \
df_operations_eu_production.loc[df_operations_eu_production['point_key'] \
                                .isin(['PRD-00202', 'PRD-00234', 'PRD-00240', 'PRD-00168', 'PRD-00200', 'PRD-00243']), 'value_in_bln_m3'] \
                                .apply(lambda x: x * 0)  # сбрасываю дубли по Германии

df_operations_eu_production['country_name'] = ''

for index, row in df_operations_eu_production.iterrows():
    if row['adjacent_country'] in list(countries_dict.keys()):
        df_operations_eu_production.at[index,'country_name'] = countries_dict[row['adjacent_country']]
    else:
        df_operations_eu_production.at[index,'country_name'] = ''

df_operations_eu_production[['year','country_name','value_in_bln_m3']] \
                            .groupby(by=['year','country_name']).sum()   # проверка страны-производители
```

```python
df_operations_eu_storage = df_operations.loc[df_operations['id_point_type']==6.0]  # Storage point

df_operations_eu_storage.loc[pd.isna(df_operations_eu_storage['adjacent_country']),'adjacent_country'] = df_operations_eu_storage.loc[pd.isna(df_operations_eu_storage['adjacent_country']),'operator_key'].apply(lambda x: x[:2])

df_operations_eu_storage['country_name'] = ''

for index, row in df_operations_eu_storage.iterrows():
    if row['adjacent_country'] in list(countries_dict.keys()):
        df_operations_eu_storage.at[index,'country_name'] = countries_dict[row['adjacent_country']]
    else:
        df_operations_eu_storage.at[index,'country_name'] = ''

df_operations_eu_storage['country_name'].unique() #проверка на страны с задействованными ПГХ
```

```python
df_operations_eu_consumption = df_operations.loc[df_operations['id_point_type']==18.0] # Aggregated Point - Final Consumers

df_operations_eu_consumption.loc[pd.isna(df_operations_eu_consumption['adjacent_country']),'adjacent_country'] = \
    df_operations_eu_consumption.loc[pd.isna(df_operations_eu_consumption['adjacent_country']),'operator_key'].apply(lambda x: x[:2])

df_operations_eu_consumption['country_name'] = ''

for index, row in df_operations_eu_consumption.iterrows():
    if row['adjacent_country'] in list(countries_dict.keys()):
        df_operations_eu_consumption.at[index,'country_name'] = countries_dict[row['adjacent_country']]
    else:
        df_operations_eu_consumption.at[index,'country_name'] = ''

df_operations_eu_consumption['country_name'].unique() #проверка на страны, по которым есть данные о потреблении
```

```python
df_import_overall = pd.concat([df_operations_pipe_import,
                               df_operations_lng_import])

df_import_overall = df_import_overall[['country_name', 'date', 'month', 'year', 'value_in_bln_m3']]
df_import_overall['type'] = 'import'
```

```python
df_operations_eu_production = df_operations_eu_production[['country_name', 'date', 'month', 'year', 'value_in_bln_m3']]
df_operations_eu_production['type'] = 'domestic_production'
```

```python
df_operations_eu_storage_overall = df_operations_eu_storage[['country_name', 'date', 'month', 'year', 'value_in_bln_m3']]

df_operations_eu_storage_overall.loc[:,'value_in_bln_m3'] = df_operations_eu_storage_overall.loc[:,'value_in_bln_m3'].apply(lambda x: x * -1)

df_operations_eu_storage_overall['type'] = 'storage'
```

```python
df_eu_gas = pd.concat([df_operations_eu_storage_overall,
                       df_operations_eu_production,
                       df_import_overall])

df_eu_gas['lng/pipe'] = df_eu_gas['country_name'].apply(lambda x: 'LNG' if x == 'LNG' else 'Pipeline gas')
df_eu_gas['season'] = df_eu_gas['month'].apply(lambda x: 'winter' if x in [10,11,12,1,2,3] else 'summer')

df_eu_gas.reset_index(drop=True, inplace=True)

df_eu_gas.to_csv(r'/Users/artemohotnikov/Documents/__code/gzprm/ready_2/df_eu_gas.csv', index=False)
```
